#### ECE BOARD EXAM PROBLEM

#### #1

a. Filename: Instru = [“Name”, “GEAS”, “Electronics >70”]; where track is constant as
Instrumentation and hometown Luzon

```python
board = pd.read_excel('board2.xlsx')
board
```

```python
instru = board.loc[(board['Electronics'] > 70) 
        & (board['Track'] == 'Instrumentation') 
        & (board['Hometown'] == 'Luzon'),
        ['Name', 'GEAS', 'Electronics']].reset_index(drop=True)
instru
```

b. Filename: Mindy = [ “Name”, “Track”, “Electronics”, “Average >=55”]; where hometown is constant as Mindanao and gender Female

```python
average = ['Math', 'Electronics', 'GEAS', 'Communication']

board['average'] = board[average].mean(axis=1)
board
```

```python
mindy = board.loc[
    (board['average'] >= 55) &
    (board['Gender'] == 'Female') &
    (board['Hometown'] == 'Mindanao'),
    ['Name', 'Track', 'Electronics', 'average']].reset_index(drop=True)
mindy
```

#### #2

Create a visualization that shows how the different features contributes to average grade. Does
chosen track in college, gender, or hometown contributes to a higher average score?

```python
import matplotlib.pyplot as plt
```

### Average

```python
fig1 = plt.figure(figsize=(20, 4))
colors = ['green']
plt.bar(board['Name'], board['average'], color = colors)
plt.title('Average Grade of Students')
```

### Hometown

```python
fig2 = plt.figure(figsize=(20, 4))
colors = ['green']
plt.bar(board['Hometown'], board['average'], color = colors)
plt.title('Hometown of Students')
```

Referring to the bar graph, we can conclude that living in Luzon has more advantage in the board exams than students living in Visayas and Mindanao.

### Gender

```python
fig3 = plt.figure(figsize=(20, 4))
colors = ['green']
plt.bar(board['Gender'], board['average'], color = colors)
plt.title('Gender of Students')
```

Based on the bar graph, we can conclude that more female students have higher average than male students taking the board exam.

### Track

```python
fig4 = plt.figure(figsize=(20, 4))
colors = ['green']
plt.bar(board['Track'], board['average'], color = colors)
plt.title('College Track of Students')
```

According to the bar graph shown, we can say that students who took the Microelectronics track have more advantage in the board exam than students who took instrumentation and Communication.
